```python
import pandas as pd
```

# BOULDER

```python
#open bikedata 2017
df1=pd.read_csv('BlockGroup/Bike Data/Boulder2017.csv')
df1.head()
```

```python
#open bikelocation 2017
df2=pd.read_csv('BlockGroup/Bike Data/Boulder_Location.csv')
df2.head()
```

```python
#Merge
df3 = pd.merge(df2, df1, how ='left', left_on = "Checkout Station", right_on = "Checkout Station")
df3
```

```python
df4 = df3.groupby(['Checkout Station','Month']).size()
df4
```

```python
df4.to_csv('BlockGroup/Bike Data/Boulder1.csv')
```

# CHATTANOOGA

```python
#open bikedata 2017
df5=pd.read_csv('BlockGroup/Bike Data/Chattanooga2017.csv')
df5.head()
```

```python
#open bikelocation 2017
df6=pd.read_csv('BlockGroup/Bike Data/Chatanooga_Location.csv')
df6.head()
```

```python
#Merge
df7 = pd.merge(df5, df6, how ='left', left_on = "StartStationName", right_on = "Address")
df7
```

```python
df8 = df7.groupby(['StartStationName','Month']).size()
df8
```

```python
df8.to_csv('BlockGroup/Bike Data/Chattanooga1.csv')
```
